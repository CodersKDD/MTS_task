```python
# !pip install tensorflow
import torch. nn as nn
import torch.nn.functional as Fun
import tensorflow as tf
from tensorflow import keras
```

```python
x_train, x_test, y_train, y_test = train_test_split(df[{'year','mouth','day', 'moskva_pogruzka', 'moskva_razgruka'}], df["moskva_hran"], train_size=0.7, shuffle=False) #, shuffle=False
```

```python

model =  tf.keras.Sequential()
model.add(tf.keras.layers.SimpleRNN(64,return_sequences=True, input_shape=(x_train.shape[1], 1 )))
model.add(tf.keras.layers.LSTM(20,return_sequences=True))
# model.add(tf.keras.layers.Dropout(0.5))
# model.add(tf.keras.layers.LSTM(20,return_sequences=True))
model.add(tf.keras.layers.Dense(1))
model.compile(loss=tf.keras.losses.MSLE, optimizer='rmsprop')
```

```python

model.fit(x_train, y_train, batch_size=32, epochs=500, verbose="auto")
```

```python
predic = model.predict(x_test)
```
